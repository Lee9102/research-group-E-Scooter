```python
import pandas as pd
import numpy as np
import keras
```

```python
a = np.arange(10)
ds = keras.utils.timeseries_dataset_from_array(
                                          data = a[:-3],
                                          targets = a[3:],
                                          sequence_length = 3,
                                          batch_size = 2)

```

```python
b = np.arange(120)
ds100 = keras.utils.timeseries_dataset_from_array(
    data = b[:-24],
    targets = b[24:],
    sequence_length = 24,
    batch_size = 7,
    sampling_rate = 2
)
```

```python
def show_dataset(ds):
    for inputs, targets in ds:
        for i in range(inputs.shape[0]):
            print([int(x) for x in inputs[i]], int(targets[i]))
```

```python
show_dataset(ds100)
```

```python
for inputs, targets in ds100:
    print(inputs.shape)
```
